```python
from langchain_community.document_loaders import WebBaseLoader
loader=WebBaseLoader("https://careers.cognizant.com/india-en/jobs/00057579891/manager-projects/")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
page_data
```

```python
from langchain_core.prompts import PromptTemplate
promt_extract=PromptTemplate.from_template(
    """
    ### ScraPED TEXT FROM WEBSITE:
    {page_data}
    ###INSTRUCTION:
    The scraped text is fron the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing
    following keys: 'role','experience','skills' and 'descripti0n'.
    Only return the valid JSON.
    ###VALID JSON (NO PREAMBLE):
    """
)
chain_extract=promt_extract | llm
res=chain_extract.invoke(input={'page_data':page_data})
type(res.content)  
```

```python
res.content
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser=JsonOutputParser()
json_res=json_parser.parse(res.content)
json_res
```
